```
#1st Solution

#Making imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(6.0,6.0)


#data preprocessing

my_data=[[-2,-1],[1,1],[3,2]]
data=pd.DataFrame(my_data)
X=data.iloc[:,0]
y=data.iloc[:,1]
plt.scatter(X,y)
plt.show()


#building model

X_mean=np.mean(X)
y_mean=np.mean(y)


#to find slope m

num=0
den=0
for i in range(len(X)):
  num+=(X[i]-X_mean)*(y[i]-y_mean)
  
  den+=((X[i]-X_mean)**2)
  
m=num/den
c=y_mean-m*X_mean

print("Slope=",m,",","y_intercept=",c)


#prediction making

y_pred=m*X+c

plt.scatter(X,y)
plt.scatter(X,y_pred,color='red')
plt.plot(X,y_pred,color='green')
plt.show()
```
