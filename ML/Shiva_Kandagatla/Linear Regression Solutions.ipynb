```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def predict(X,intercept,slope):
	predictions = list()
	for i in range(len(X)):
		pred = intercept + slope * X[i]
		predictions.append(pred)
	return predictions

def mean(values):
	return sum(values) / float(len(values))
 
def numerator(x, mean_x, y, mean_y):
	num = 0.0
	for i in range(len(x)):
		num += (x[i] - mean_x) * (y[i] - mean_y)
	return num
    
def denominator(values, mean):
	return sum([(x-mean)**2 for x in values])
 
def coefficients(x,y):
	x_mean, y_mean = mean(x), mean(y)
	slop = numerator(x, x_mean, y, y_mean) / denominator(x, x_mean)
	intercpt = y_mean - slop * x_mean
	return [intercpt, slop]
```

Problem 1: 
	Consider the following set of points: {(-2 , -1) , (1 , 1) , (3 , 2)} 
	a) Find the least square regression line for the given data points. 
	b) Plot the given points and the regression line in the same rectangular system of axes. 

```python
dataset = pd.read_csv('./Problem 1/DataSet_Problem1.csv')
X = dataset.iloc[ : , 0].values
y = dataset.iloc[ : , 1].values

intercept, slope = coefficients(X,y)
ypred = predict(X,intercept,slope)
print(f"The Least Square Regression Line for given DataSet is Given by :  y = {slope} * x + {intercept}")
```

```python
print("Visualizing the Data Points along with the Regression Line")

plt.scatter(X,y, color='red')
plt.plot(X,ypred, color='blue')
plt.title("X vs y")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

Problem 3:
	The values of y and their corresponding values of y are shown in the table below 
	x  0  1  2  3  4
	y  2  3  5  4  6
    
a) Find the least square regression line y = a x + b. 
b) Estimate the value of y when x = 10.

```python
dataset = pd.read_csv('./Problem 3/DataSet_Problem3.csv')
X = dataset.iloc[ : , 0].values
y = dataset.iloc[ : , 1].values

intercept, slope = coefficients(X,y)
ypred = predict(X,intercept,slope)
print(f"The Least Square Regression Line for given DataSet is Given by :  y = {slope} * x + {intercept}")
```

```python
print("Visualizing the Data Points along with the Regression Line")

plt.scatter(X,y, color='red')
plt.plot(X,ypred, color='blue')
plt.title("X vs y")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
print("Estimating y value when x = 10 ....")
estimation = (slope * 10) + intercept
print(f"Value of y when x = 10 is \'{estimation}\'")
```

Problem 4:
	The sales of a company (in million dollars) for each year are shown in the table below. 
	x (year)  2005  2006  2007  2008  2009
	y (sales)  12  19  29  37  45

a) Find the least square regression line y = a x + b. 
b) Use the least squares regression line as a model to estimate the sales of the company in 2012.

```python
dataset = pd.read_csv('./Problem 4/DataSet_Problem4.csv')
X = dataset.iloc[ : , 0].values
y = dataset.iloc[ : , 1].values

intercept, slope = coefficients(X,y)
ypred = predict(X,intercept,slope)
print(f"The Least Square Regression Line for given DataSet is Given by :  y = {slope} * x + {intercept}")
```

```python
print("Visualizing the Data Points along with the Regression Line")

plt.scatter(X,y, color='red')
plt.plot(X,ypred, color='blue')
plt.title("Year vs Sales")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
print("Estimating Sales in the year 2012........ ")
estimation = (slope * 2012) + intercept
print(f"Estimated Sales in the year 2012 is \'{estimation}\'")
```
